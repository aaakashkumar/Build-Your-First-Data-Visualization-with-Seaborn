# Data Visualization - What is it and Why Do We Need It?

## Data Visualization
Data visualization is the representation of data in a pictorial or graphical format

# Why Is Data Visualization Important?
We need data visualization because a  visual summary of information makes it easier to identify patterns and trends rather than looking through thousands of rows on a spreadsheet.
* Visualizing complex data through charts and graphs is easier than poring over spreadsheets
* Easier to gain insights when data is visualized
* Visualization also makes communicating insights easier

# Introduction to Seaborn

Seaborn is basically built on top of Matplotlib. Matplotlib has proven to be an incredibly useful and popular visualization tool, but there are some aspects of it that often leaves much to be desired. Seaborn is complimentary to Matplotlib and specifically targets statistical data visualization.

**Seaborn**
* Seaborn targets statistical visualization, provides more variety and generally uses fewer syntax
* Seaborn’s default parameters and themes are both more interesting and visually pleasing
* Seaborn provides an API on top of Matplotlib and avoids a ton of boilerplate by providing default themes which are commonly used.

**Matplotlib**
* Matplotlib is generally used for basic data visualization and is typically more wordy
* Matplotlib’s default parameters are not very aesthetically pleasing
* Matplotlib’s API is low level, so performing sophisticated visualization requires lot of boilerplate code

# Visualizing Data with Seaborn

## Loading and Viewing the Dataset

```
# importing the dataset

athlete_data = pd.read_csv("athletes.csv")
country_data = pd.read_csv("countries.csv")

tips_data = sns.load_dataset("tips")
```

## Visualizing Statistical Relationships

### What is the relationship between population and GDP?

```
sns.relplot(x = 'population', y = 'gdp_per_capita', data = country_data)
plt.show()
```

### What is the relationship between height and weight among the athletes?

```
sns.relplot(x = 'height', y = 'weight', data = athlete_data)
plt.show()
```

### Height and Weight correlation based on sex

```
sns.relplot(x = 'height', y = 'weight', hue = 'sex', data = athlete_data)
plt.show()
```

### Height and Weight correlation based on sport

```
sns.relplot(x = 'height', y = 'weight', hue = 'sport', data = athlete_data)
plt.plot()
```

## Plotting categorical data

### What is the connection between tip amount and day of week?

```
sns.catplot(x = 'day', y = 'tip', data = tips_data)
```

### Swarm plot

```
sns.catplot(x = 'day', y = 'tip', kind = 'swarm', data = tips_data)
```

### Who typically gets higher tips – males or females?

```
sns.catplot(x = 'day', y = 'tip', kind = 'swarm', hue = 'sex', data = tips_data)
```

### Boxplot

Understand the five number summary here: <br>
https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/box-plot-review

Interquartile range: <br>
Q<sub>3</sub> - Q<sub>1</sub>

Anything outside 1.5 multiplied by the interquartile range is considered an outlier in both the positive and negative directions.

```
sns.catplot(x = 'day', y = 'tip', kind = 'box', data = tips_data)
```

```
sns.catplot(x = 'day', y = 'tip', kind = 'box', hue = 'sex', data = tips_data)
```

### Violin plots

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. Violin plots have many of the same summary statistics as box pots, the white dot represents the median, the thick black bar in the center represents the interquatile range, and the thin grey line represents the rest of the distribution except for points are determined to be outliers. On each side of the gray line is a kernel density estimation to show the distribution shape of the data. Wider sections of the violin plot represent the higher probability that members of the population will take on the given value. The skinniest sections represent a lower probability. 

```
sns.catplot(x = 'day', y = 'tip', kind = 'violin', hue = 'sex', data = tips_data)
```

### What are the types of sports and what does their participation look like?

```
sns.countplot(x = athlete_data['sport'], order = athlete_data['sport'].value_counts().index)
```

## Vizualizing the Distribution of a Dataset

### What does the total bill look like in the restaurant?

```
sns.distplot(tips_data['total_bill'], kde=False)
```

### Kernel Density Estimate (KDE) Plot

```
sns.kdeplot(tips_data['total_bill'], shade=True)
```

### What is the relationship between total_bill and tip?

<font color = "#009900">Bivariate Plot - 2D KDE Plot</font>

```
sns.jointplot(x = tips_data['total_bill'], y = tips_data['tip'], kind = 'kde')
```
